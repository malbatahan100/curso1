```
USE Northwind
GO
IF EXISTS
(
	SELECT *
	FROM sysobjects
	WHERE
			xtype = 'FN'
		AND
			name = 'ufn_categoria'
)
DROP FUNCTION ufn_categoria
GO
CREATE FUNCTION ufn_categoria
(
	@CategoryID INT
)
RETURNS NVARCHAR(15)
AS
BEGIN
	RETURN
	(
		SELECT [CategoryName]
		FROM [dbo].[Categories]
		WHERE [CategoryID] = @CategoryID
	)
END
GO
SELECT [ProductID]
      ,[ProductName]
      ,[SupplierID]
      ,[CategoryID]
	  ,dbo.ufn_categoria([CategoryID]) AS Categoría
      ,[QuantityPerUnit]
      ,[UnitPrice]
      ,[UnitsInStock]
      ,[UnitsOnOrder]
      ,[ReorderLevel]
      ,[Discontinued]
  FROM [dbo].[Products]
GO
```

Las funciones recursivas en SQL Server son un tipo de función que se llama a sí misma para resolver un problema.

Se utilizan para realizar cálculos que requieren una repetición de operaciones, como la búsqueda de jerarquías, el cálculo de factoriales, o la exploración de estructuras de datos complejas como árboles.



```
-- ================================================
-- Template generated from Template Explorer using:
-- Create Scalar Function (New Menu).SQL
--
-- Use the Specify Values for Template Parameters
-- command (Ctrl-Shift-M) to fill in the parameter
-- values below.
--
-- This block of comments will not be included in
-- the definition of the function.
-- ================================================
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		Edgard Lucho
-- Create date: 30/10/2024
-- Description:	Función para calcular el factorial de un número entero positvo
-- =============================================
CREATE OR ALTER FUNCTION Factorial
(
	@n int
)
RETURNS int
AS
BEGIN
	DECLARE @f INT=1
	IF @n = 0 OR @n = 1
		SET @f = 1;
	ELSE
		SET @f = (@n * dbo.Factorial(@n-1));
	RETURN @f;
END
GO
SELECT dbo.Factorial(3) AS Factorial;
GO
```





```
USE Northwind
GO
CREATE OR ALTER FUNCTION Factorial
(
	@n int
)
RETURNS int
AS
BEGIN
	RETURN IIF(@n = 0 OR @n = 1, 1, (@n * dbo.Factorial(@n-1)));
END
GO
SELECT dbo.Factorial(6) AS Factorial;
GO
```

```
-- ================================================
-- Template generated from Template Explorer using:
-- Create Multi-Statement Function (New Menu).SQL
--
-- Use the Specify Values for Template Parameters
-- command (Ctrl-Shift-M) to fill in the parameter
-- values below.
--
-- This block of comments will not be included in
-- the definition of the function.
-- ================================================
SET ANSI_NULLS ON
GO
SET QUOTED_IDENTIFIER ON
GO
-- =============================================
-- Author:		<Author,,Name>
-- Create date: <Create Date,,>
-- Description:	<Description,,>
-- =============================================
CREATE OR ALTER FUNCTION utf_productos_x_codigo_categoria
(
	@CategoryID INT
)
RETURNS @T TABLE
(
	[ProductID] [int]  NOT NULL,
	[ProductName] [nvarchar](40) NOT NULL,
	[Proveedor] [nvarchar](15) NULL,
	[Categoria] [nvarchar](15) NULL,
	[QuantityPerUnit] [nvarchar](20) NULL,
	[UnitPrice] [money] NULL,
	[UnitsInStock] [smallint] NULL,
	[UnitsOnOrder] [smallint] NULL,
	[ReorderLevel] [smallint] NULL,
	[Discontinued] [bit] NOT NULL
)
AS
BEGIN
	INSERT INTO @T
	(
		[ProductID]
		,[ProductName]
		,[Proveedor]
		,[Categoria]
		,[QuantityPerUnit]
		,[UnitPrice]
		,[UnitsInStock]
		,[UnitsOnOrder]
		,[ReorderLevel]
		,[Discontinued]
	)
	SELECT
		[ProductID]
		,[ProductName]
		,dbo.ufn_proveedor([SupplierID]) AS Proveedor
		,dbo.ufn_categoria([CategoryID]) AS Categoria
		,[QuantityPerUnit]
		,[UnitPrice]
		,[UnitsInStock]
		,[UnitsOnOrder]
		,[ReorderLevel]
		,[Discontinued]
	FROM Products
	WHERE CategoryID = @CategoryID;
	RETURN
END
GO
SELECT * FROM dbo.utf_productos_x_codigo_categoria(6);
GO
```
